## $\color{orange}{\text{Introducción}}$

La secretaría distrital de salud (SDS) es la organización responsable de garantizar el ejercicio efectivo del derecho a la salud de toda la población de Bogotá. En este orden de ideas, la SDS ha sido la organización encargada de establecer medidas de control y mitigación del virus *SARS-CoV-2* en la ciudad desde su llegada el 6 de marzo de 2020 en un vuelo proveniente de Milán, y que en la actualidad ha reportado más de $1442\;000$ casos en la ciudad. En la toma de medidas se han hecho intervenciones farmacológicas (IF) como protocolos de bioseguridad, la ampliación de recursos sanitarios en la ciudad y estrategias de vacunación, sin embargo, debido a la limitación de no poder expandir indefinidamente los recursos sanitarios, se ha alcanzado un alto porcentaje de ocupación (del orden de $94\%$ en la segunda ola vivida en la ciudad en el mes de enero de $2021$), generando un alto estrés en el sistema de salud en largos periodos de tiempo. Esto ha conllevado a restricciones totales en la movilidad y cuarentenas sectorizadas a nivel de localidades y UPZ, así como toques de queda y pico y cédula generalizado. Luego, se hace necesario poder cuantificar el potencial impacto de diferentes IF o intervenciones no farmacológicas (INF) como el uso de tapabocas, cuarentenas sectorizadas y restricciones de movilidad. Por lo demás, actualmente la ciudad se encuentra finalizando una tercera ola de *SARS-CoV-2* que llegó a saturar los recursos sanitarios de la ciudad instalados hoy en día.

El secretario distrital de salud Alejandro Gómez López junto a la subsecretaría distrital de salud María Clemencia Mayorga Rámirez manejan todos los procesos de la SDS, dentro de estos se encuentra la Subsecretaría de Salud Pública - Dirección Epidemiología, Análisis y Gestión de Políticas de Salud Colectiva - Subdirección de Vigilancia en Salud Pública (*SSP-DEAGPSC-SVSP*) donde pertenece el equipo de Ánalisis de Situación de Salud (*ASIS*). En el equipo ASIS es precisamente donde se desenvuelve el equipo de modelación matemática conformada por físicos, matemáticos y estadísticos (*ASIS-FEM*). Este equipo se concentra en hacer proyecciones de potenciales escenarios del rumbo de la epidemia en la ciudad de Bogotá para la ocupación de UCI utilizando un modelo de ecuaciones diferenciales ordinarias no lineales acopladas denominado [Squire](https://mrc-ide.github.io/squire/index.html) o, más recientemente, un modelo basado en agentes denominado [FRED (A Framework for Reconstructing Epidemic Dynamics)](https://doi.org/10.1186/1471-2458-13-940). ## $\color{orange}{\text{Exploración de datos y verificación de calidad}}$

Finalmente, en esta descripción de los datos vale la pena resaltar la siguiente figura que nos muestra al igual que el tablero de SALUDATA, la ocupación de hospitalización general en términos absolutos y en términos porcentuales. Además, esto corroborá la hipótesis inicial del tercer pico de la pandemia vivido en la ciudad.

```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime

d0 = date(2020, 5, 20)
datelist = pd.date_range(start = d0, end = date.today(), periods = df['% Ocupación'].size) # Fechas

df = df.reset_index()
df = df.reset_index()

df['CamasA'] = np.where(((df['index']>=378) & (df['index']<=404)) | (df['index']>=431),
                          df['Camas Ocupadas COVID 19'],
                          df['Camas asignadas COVID 19']) # Camas Asignadas

df['CamasO'] = np.where(((df['index']>=378) & (df['index']<=404)) | (df['index']>=431),
                          df['Camas asignadas COVID 19'],
                          df['Camas Ocupadas COVID 19']) # Camas Ocupadas

df['% Ocupación'] = np.where(((df['index']>=378) & (df['index']<=404)) | (df['index']>=431),
                               df['Camas asignadas COVID 19']/df['Camas Ocupadas COVID 19']*100,
                               df['Camas Ocupadas COVID 19']/df['Camas asignadas COVID 19']*100) # Porcentaje de ocupación

df['Camas asignadas COVID 19'] = df['CamasA']
df['Camas Ocupadas COVID 19'] = df['CamasO']

plt.figure(figsize=(21,6))
plt.subplot(1,2,1)
plt.plot_date(datelist, df['Camas Ocupadas COVID 19'],color="green",marker="o",linestyle="-",label='Camas ocupadas')
plt.plot_date(datelist, df['Camas asignadas COVID 19'],color="red",marker="o",linestyle="-",label='Camas disponibles')
plt.xlabel('Fecha')
plt.xticks(pd.date_range(start = d0, end = date.today(), periods = 12), rotation='vertical')
plt.yticks(np.linspace(0,3000,11))
plt.ylabel('Camas ocupadas')
plt.legend()
plt.grid(b=True, which='major', axis='both')
plt.title('Camas ocupadas en hospitalización general COVID 19')
plt.subplot(1,2,2)
plt.plot_date(datelist, df['% Ocupación'],color="blue",marker="o",linestyle="-")
plt.xlabel('Fecha')
plt.xticks(pd.date_range(start = d0, end = date.today(), periods = 12), rotation='vertical')
plt.yticks(np.linspace(30,100,8))
plt.ylabel('% Ocupación')
plt.grid(b=True, which='major', axis='both')
plt.title('Porcentaje de ocupación en hospitalización general COVID 19')
plt.show()
```

Por último, se resalta que en este caso específico no hay pérdida de datos, errores tipográficos y algunas inconsistencias en la codificación por una posible carga errónea de los datos durante los días de mayo.

## $\color{orange}{\text{Motivación del proyecto de Machine Learning}}$

Hasta el momento se ha presentado en un contexto general un problema que es de bastante interés por su actualidad que corresponde al *estudio de recursos sanitarios de hospitalización general generados por la pandemia de SARS-CoV-2*. También es claro la importancia que tiene para un gobierno poder brindar bienestar y salud a sus ciudadanos así como poder atenderlos de manera oportuna. Esto conlleva a plantearse un tipo adicional de preguntas posteriores a la toma de decisiones, que en este caso corresponde a las medidas de mitigación no farmacológicas tomadas en la ciudad, más precisamente, a preguntas del tipo qué desempeño han tenido las decisiones tomadas en Bogotá en comparación a otras ciudades del país o si han sido efectivas y oportunas las medidas de mitigación tomadas en la ciudad. Por esta razón es importante implementar no sólo medidas que permitan saber la capacidad necesario de recursos sanitarios para satisfacer las demandas que tenga la ciudad sino también medidas de desempeño del problema de analítica de datos. Para este fin, es bastante útil ver la correlación entre el problema de analítica de datos y el aprendizaje de máquina o machine learning.

Por otro lado, dado la situación mundial actual generada a nivel económico por el SARS-CoV-2/Covid-19, es claro que las personas cada vez se han agotado de las medidas de mitigación como cuarentenas sectorizadas y que las potenciales medidas de mitigación de la pandemia se concentran en la vacunación masiva de la población, instalación de puntos de tomas de muestras y también la detección de brotes o alertas tempranas que sirvan a mantener en orden la ciudad. Ante esta necesidad, es claro que los modelos matemáticos utilizados deben ya no sólo concentrar su preocupación en la proyección de potenciales escenarios de reapertura sino la desagregación de los individuos a nivel espacial que permita tomar medidas de forma focalizada que no interrumpan el desarrollo económico bastante deteriorado en la ciudad luego de más de un año de pandemia. En este orden de ideas, se propone implementar un modelo de agentes denominado [FRED (A Framework for Reconstructing Epidemic Dynamics)](https://doi.org/10.1186/1471-2458-13-940).

El machine learning o aprendizaje de máquina (ML) establece un aprendizaje por diferentes modelos de la computadora a partir del uso de datos (entrenamiento), lo cual permite establecer un proceso de predicciones o proyecciones una vez se tiene bien implementado el mecanismo. En particular, para el ámbito de la pandemia se han establecido modelos de ML que permitan reducir el riesgo de mortalidad mediante la aplicación efectiva de aislamiento y rastreo de casos sospechosos o de cuarentenas sectorizadas, un ejemplo de ello esta en el artículo de Mahdavi et. al., *[A machine learning based exploration of COVID-19 mortality risk](https://doi.org/10.1371/journal.pone.0252384)*, o también proyectar las curvas epidemicas de casos confirmados, hospitalizaciones y fallecimientos como se muestra en el trabajo de Zoabi et. al., [Machine learning-based prediction of COVID-19 diagnosis based on symptoms](https://doi.org/10.1038/s41746-020-00372-6). Así, existe un amplio espectro de aplicación del ML en modelos matemáticos enfocados a la epidemiología y que para fines prácticos se lleva al siguiente problema que es bien conocido y complejo en el contexto de modelos dinámicos: dado un conjunto de parámetros de entrada en el modelo matemático y un conjunto de datos de entrada (en este caso la serie histórica de hospitalizados en Bogotá y su porcentaje de ocupación así como la serie de tiempo de fallecimientos diarios, ocupación UCI, entre otros), cuáles son los parámetros de ajuste óptimo para que el modelo reproduzca la tendencia que han tenido los datos. En otras palabras, cuáles son los parámetros óptimos para calibrar el modelo dado un conjunto de datos de entrada. Por lo tanto, se hace bastante evidente que el problema a abordar con ML es la calibración del ML de modo que las proyecciones del modelo se puedan hacer de la forma más automática posible y que sólo requiera una supervisación del experto en ciencia de datos.

## $\color{orange}{\text{Definición del problema}}$

**Salida del problema de ML**

Se espera que el problema de ML proporcione los parámetros de tiempos de estancia entre diferentes compartimentos en el modelo matemático implementado y que a diferencia del modelo Squire ya implementado en la secretaría de Salud, este nuevo modelo pueda dar información desagregada a nivel espacial como puede ser a nivel de localidad o UPZ en Bogotá. Para este fin, se propone la estimación de parámetros mediante un muestreo de parámetros.

**Datos de entrada del problema de ML**

Las variables de entrada o los datos de entrenamiento para tener en cuenta son:

* Serie histórica de casos hospitalizados y porcentaje de ocupación en hospitalización general para pacientes diagnosticados con SARS-CoV-2/Covid-19 en Bogotá. A la fecha son del orden de 475 datos a los que se tiene acceso.
* Serie histórica de casos en UCI y porcentaje de ocupación en UCI para pacientes diagnosticados con SARS-CoV-2/Covid-19 en Bogotá. A la fecha son del orden de 507 datos a los que se tiene acceso.
* Serie histórica de casos fallecidos por fecha de muerte para pacientes que fueron diagnosticados con SARS-CoV-2/Covid-19 en Bogotá. A la fecha son del orden de 524 datos a los que se tiene acceso.

Aquí vale la pena aclarar que estos datos son de actualización diaria y por temas de confidencialidad solo se muestra un análisis similar al que se hizo con los datos de hospitalización para el caso de UCIs. Además, todos los datos son de importancia para el problema para establacer de forma óptima los parámetros de ajuste del modelo.

```python
import requests
import pandas as pd
import numpy as np
from datetime import date
import openpyxl

# Diferencia de fechas a partir del 15 de mayo de 2020
d0 = date(2020, 4, 8)
d1 = date.today()
delta = d1 - d0
# URL de los datos de SALUDATA
url = 'https://datosabiertos.bogota.gov.co/dataset/adebac34-7974-4aa8-a31d-97b2b6796e4b/resource/6657cbbc-8277-4caf-8823-3350f253ae64/download/osb_ocupacion_ucis_covid-19_01082021.csv'
# Importación de datos en Pandas
df = pd.read_csv(url, index_col=0, encoding='latin-1', error_bad_lines=False, sep=';',
                 nrows=delta.days-1, usecols = [0,1,2,3])
df['Camas UCI asignadas COVID 19'] = np.where(df['Total camas UCI COVID 19 reportadas por IPS']<10,
                                              df['Total camas UCI COVID 19 reportadas por IPS']*1000,
                                              df['Total camas UCI COVID 19 reportadas por IPS'])
df['Camas UCI ocupadas Covid-19'] = np.where(df['Camas UCI ocupadas Covid-19']<10,
                                             df['Camas UCI ocupadas Covid-19']*1000,
                                             df['Camas UCI ocupadas Covid-19'])
df.tail(10)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime

d0 = date(2020, 4, 8)
datelist = pd.date_range(start = d0, end = date.today(), periods = df['Ocupación UCI COVID 19'].size) # Fechas

df['% Ocupación'] = df['Camas UCI ocupadas Covid-19']/df['Camas UCI asignadas COVID 19']*100 # Porcentaje de ocupación

plt.figure(figsize=(21,6))
plt.subplot(1,2,1)
plt.plot_date(datelist, df['Camas UCI asignadas COVID 19'],color="green",marker="o",linestyle="-",label='Camas disponibles')
plt.plot_date(datelist, df['Camas UCI ocupadas Covid-19'],color="red",marker="o",linestyle="-",label='Camas ocupadas')
plt.xlabel('Fecha')
plt.xticks(pd.date_range(start = d0, end = date.today(), periods = 12), rotation='vertical')
plt.yticks(np.linspace(0,2500,11))
plt.ylabel('Camas ocupadas')
plt.legend()
plt.grid(b=True, which='major', axis='both')
plt.title('Camas ocupadas en UCI COVID 19')
plt.subplot(1,2,2)
plt.plot_date(datelist, df['% Ocupación'],color="blue",marker="o",linestyle="-")
plt.xlabel('Fecha')
plt.xticks(pd.date_range(start = d0, end = date.today(), periods = 12), rotation='vertical')
plt.yticks(np.linspace(30,100,8))
plt.ylabel('% Ocupación')
plt.grid(b=True, which='major', axis='both')
plt.title('Porcentaje de ocupación en UCI COVID 19')
plt.show()
```

## $\color{orange}{\text{Medidas de desempeño}}$

**Marco de trabajo**

En el contexto del muestreo de parámetros por lo general se utiliza un espacio de parámetros $\Omega\in\mathbb{R}^{n}$ para $n$ parámetros escalares y que para fines metodológicos se usa el muestreo de hipercubo latino ([Latin hypercube sampling](https://dx.doi.org/10.2307%2F2291282)). En este espacio de parámetros se asume la existencia de distribuciones de probabilidad específicas para muestrear aleatoriamente los parámetros y estimar los valores óptimos a partir de una estimación por máxima verosimilitud (MLE). En particular, se asume que los tiempos de estancia siguen una distribución binomial negativa.

**Medida de desempeño**

Con lo expuesto anteriormente, se establece como medida de desempeño, la verosimilitud obtenida del muestreo de parámetros tal que luego de $N$ simulaciones, el valor con mayor verosimilitud es el que mejor ajusta el conjunto de parámetros.

**Posibles errores y comentarios sobre la solución del algoritmo**

Ahora bien, es posible pensar que existan falsos positivos (FP) o falsos negativos (FN) en este proceso. Sin embargo, dado que todos los parámetros no solo se contrastan al momento de comparar los datos de entrenamiento sino que también se compararon con la literatura cuando se construye el modelo, podemos tener certeza que estos casos son casos bastante atípicos. Aquí vale la pena mencionar que para estimar si la solución obtenida es una buena solución o no, siempre se estima un rezago de 14 días en los datos de fallecidos (que pueden deberse a múltiples causas como la demora en el reporte de fecha de muerte para casos fallecidos o el tiempo de espera para el resultado por laboratorio de una muestra aún cuando la persona ya haya fallecido) los cuales sirven para saber si la calibración del modelo logra recuperar la tendencia adicional de las últimas dos semanas.

## $\color{orange}{\text{Partición de los datos}}$

**Consideraciones**

Para el conjunto de datos de entrenamiento que corresponde a múltiples series de tiempo, se debe tener en cuenta que los datos se asumen sin sesgo pero si con un imbalance en el sentido que en un momento de alto estres sobre el sistema, se vuelven más probables los datos que habitualmente pueden ser atípicos como una ocupación de UCI superior al 90% en Bogotá.

**Estrategia de partición**

Para hacer la partición de los datos de entrenamiento se utiliza el método de validación cruzada dejando uno fuera *(Leave-one-out cross-validation (LOOCV))*. Sin embargo, a diferencia de como es usual en este método de fijar múltiples iteraciones de una muestra de datos de prueba y el resto como datos de entrenamiento, se asume una iteración sobre los datos durante las últimas tres semanas (21 días). Para ello, se toma como conjunto de entrenamiento todos los datos con fecha anterior a las últimas tres semanas y como datos de prueba los últimos 21 datos de las series de tiempo utilizadas. Luego, se repite el ejercicio con los últimos 20 datos de la serie de tiempo y así sucesivamente hasta que queden 7 datos de la serie de tiempo. Esto con el fin de subestimar o sobrestimar los parámetros de calibración del modelo debido al rezago en los datos de fallecidos durante las últimas dos semanas.
